```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# Target search for Parkinson

target=new_client.target
target_query=target.search('CDK5')
targets=pd.DataFrame.from_dict(target_query)
targets
```

```python
# Filter targets for homo-sapiens

selected_targets=targets[targets['organism']=='Homo sapiens']
selected_targets
```

```python
#select the target from the above

selected_target=targets.target_chembl_id[1]
selected_target
```

```python
# retrieve bioactivity data for PD from CHEMBL5169188

activity=new_client.activity
res= activity.filter(target_chembl_id=selected_target).filter(stardard_type='IC50')
res
```

```python
df=pd.DataFrame.from_dict(res)
df
```

```python
# Filter for inhibitors

inhib=df[df['assay_description'].str.contains('nhibit')]
inhib
```

```python
# Select columns of interest

sel_col=['molecule_chembl_id','canonical_smiles','standard_value','standard_units','assay_description']
inhib_df=inhib[sel_col]
```

```python
# Rename the columns

inhibitor_df=inhib_df.rename(columns={'molecule_chembl_id':'ChEMBL_ID','standard_value':'IC50','standard_units':'IC50_units','assay_description':'Assay_Description'})
inhibitor_df
```

```python
# Save the data to a CSV file"""

inhibitor_df.to_csv("LRKK2_all_Inhibitors.csv", index=False)
```
